```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
playing = True

class Card:
    def __init__(self, rank, suit):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return f'{self.rank} of {self.suit}'
    
class Deck():
    
    def __init__(self):
        self.deck = []
        
        for s in suits:
            for r in ranks:
                new_card = Card(r,s)
                self.deck.append(new_card)
                
    def __str__(self):
        cards = ''
        counter = 0
        for x in self.deck:
            cards += x.__str__() + '\n'
            counter+=1
        return f'{cards}Total Cards: {counter}'
    
    def shuffle(self):
        random.shuffle(self.deck)
    
    def deal(self):
        single_card = self.deck.pop(0)
        return single_card
    
class Hand():
    
    def __init__(self):
        self.value = 0
        self.current_hand = []
        self.aces = 0
    
    def add_cards(self,card):
        self.current_hand.append(card)
        self.value += values[card.rank]
        if card.rank == "Ace":
            self.aces += 1
        
    def adjust_for_ace(self):
      if self.value > 21 and self.aces >0:
         self.value -= 10
         self.aces -= 1
                
        
class Chips:
    
    def __init__(self):
        self.total = 100  
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    
    def lose_bet(self):
        self.total -= self.bet
        

def take_bet(chips):
    while True:
        try:
            chips.bet = int(input(f"You have {chips.total} chips. How much do you want to bet?"))
        except:
            print("Please enter a number.")
        else:
            if(chips.bet <= chips.total):
                break

def hit(deck,hand):
    hand.add_cards(deck.deal())
    hand.adjust_for_ace()
        
def hit_or_stand(deck, hand):
    global playing
    acceptable = ["hit","stand"]
    decision = ""
    while decision not in acceptable:
        decision = input("Would you like to hit or stand?").lower()
        if decision == "hit":
            hit(deck,hand)
        if decision == "stand":
            print("Player stands, Dealer's turn")
            playing = False  

def show_some(player,dealer):
    print("Dealer's hand is")
    print("<<hidden card>> \n",dealer.current_hand[1])
    print("Player's hand is")
    print(*player.current_hand, sep='\n')
    
def show_all(player,dealer):
    print("Dealer's hand is")
    print(*dealer.current_hand, sep='\n')
    print("Player's hand is")
    print(*player.current_hand, sep='\n')
        
def player_busts(player,dealer,chips):
    print("Player busts!")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("Player wins!")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("Dealer busts!")
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print("Dealer wins!")
    chips.lose_bet()
    
def push(player,dealer):
    print("Dealer and Player tie! It's a push.")
    
```
